```python
import sys
import os

import pandas as pd
 
# Getting the name of the directory where the this file is present.
current = os.path.dirname(os.path.realpath('products'))
 
# Getting the parent directory name where the current directory is present.
parent = os.path.dirname(current)
 
# Adding the parent directory to the sys.path.
sys.path.append(parent)

from data_extraction import DataExtractor
from data_cleaning import DataCleaning
from database_utils import DatabaseConnector
```

```python
S3_ADDRESS = 's3://data-handling-public/products.csv'

df = DataExtractor().extract_from_s3(S3_ADDRESS)
df
```

```python
df = df.drop_duplicates()
df = DataCleaning.remove_null_values(df)
df
```

```python
df['weight'].sort_values()
df['weight'].unique()
df['weight'] = DataCleaning.remove_strings_with_invald_weight_suffix(df, 'weight')
df['weight'].sort_values()
df['weight'].unique()
df['weight'] = DataCleaning.calculate_weight_multiples(df, 'weight')
df['weight'].unique()
df['weight'] = df['weight'].apply(lambda x: x.strip(' .') if x == '77g .' else x)
df['weight'] = DataCleaning.convert_product_weights_to_kg(df, 'weight')
df['weight'].unique()
df.head(1850)
```

```python
print(df.columns.tolist())
# ['Unnamed: 0', 'product_name', 'product_price', 'weight', 'category', 'EAN', 'date_added', 'uuid', 'removed', 'product_code']
```

```python
del df['Unnamed: 0']
print(df.columns.tolist())
```

```python
df['product_name'].sort_values()
# df['product_name'].unique()
```

```python
# delete rows if all of the below columns are empty because we won't know the product
df = df[((df['product_name'] != '') & (df['EAN'] != '') & (df['uuid'] != '') & (df['product_code'] != ''))]
df
```

```python
df['product_price'].sort_values()
df['product_price'].unique()
df['product_price'] = DataCleaning.remove_alphanumerical_values(df, 'product_price')
df['product_price'].sort_values()
```

```python
df['category'].sort_values()
df['category'] = DataCleaning.remove_alphanumerical_values(df, 'category')
df['category'].sort_values()
df['category'].unique()
```

```python
df['EAN'].sort_values()
df['EAN'].unique()
```

```python
df['date_added'].sort_values()
df['date_added'].unique()
df['date_added'] = DataCleaning.remove_non_datetime_values(df, 'date_added')
df['date_added'] = DataCleaning.format_datetime_values(df, 'date_added')
df = DataCleaning.remove_null_values(df)
df['date_added'].sort_values()
```

```python
df['uuid'].sort_values()
df['uuid'].unique()
df['uuid'] = DataCleaning.remove_alphanumerical_values(df, 'uuid')
df['uuid'].unique()
```

```python
df['removed'].sort_values()
df['removed'].unique()
df['removed'] = DataCleaning.remove_invalid_product_statuses(df, 'removed')
df['removed'].unique()
```

```python
df['product_code'].sort_values()
df['product_code'].unique()
```
