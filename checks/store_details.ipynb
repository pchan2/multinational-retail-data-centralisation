```python
import sys
import os
 
# Getting the name of the directory where the this file is present.
current = os.path.dirname(os.path.realpath('stores_details'))
 
# Getting the parent directory name where the current directory is present.
parent = os.path.dirname(current)
 
# Adding the parent directory to the sys.path.
sys.path.append(parent)

from data_extraction import DataExtractor
from data_cleaning import DataCleaning
from database_utils import DatabaseConnector
```

```python
NUMBER_OF_STORES_ENDPOINT = 'https://aqj7u5id95.execute-api.eu-west-1.amazonaws.com/prod/number_stores'
number_of_stores = DataExtractor().list_number_of_stores(NUMBER_OF_STORES_ENDPOINT) # 451
```

```python
RETRIEVE_A_STORE_ENDPOINT = 'https://aqj7u5id95.execute-api.eu-west-1.amazonaws.com/prod/store_details'
df = DataExtractor().retrieve_stores_data(RETRIEVE_A_STORE_ENDPOINT, number_of_stores)
df
```

```python
df = df.drop_duplicates()
df = DataCleaning.remove_null_values(df)
df
```

```python
list(df)
# ['index', 'address', 'longitude', 'lat', 'locality', 'store_code', 'staff_numbers',
#  'opening_date', 'store_type', 'latitude', 'country_code', 'continent']
```

```python
del df['index']
df
```

```python
df['address'].sort_values()
# df['address'].unique()
```

```python
df['longitude'].unique()
df['longitude'] = DataCleaning.remove_alphanumerical_values(df, 'longitude')
df['longitude'].unique()
df['longitude'].sort_values()
```

```python
df['lat'].unique()
```

```python
df['locality'].unique()
df['locality'].sort_values()
df['locality'] = DataCleaning.remove_strings_with_numbers(df, 'locality')
df['locality'].sort_values()
df['locality'].unique()
```

```python
df['store_code'].unique()
df['store_code'] = DataCleaning.remove_values_with_non_alpha_prefix(df, 'store_code')
df['store_code'].unique()
df['store_code'].sort_values()
```

```python
df['staff_numbers'].unique()
df['staff_numbers'] = DataCleaning.remove_non_digit_values(df, 'staff_numbers')
df['staff_numbers'].unique()
```

```python
df['opening_date'].unique()
df['opening_date'] = DataCleaning.remove_non_datetime_values(df, 'opening_date')
df['opening_date'] = DataCleaning.format_datetime_values(df, 'opening_date')
df = DataCleaning.remove_null_values(df)
df['opening_date'].unique()
```

```python
df['store_type'].unique()
df['store_type'] = DataCleaning.remove_non_store_types(df, 'store_type')
df['store_type'].unique()
```

```python
df['latitude'].unique()
df['latitude'] = DataCleaning.remove_alphanumerical_values(df, 'latitude')
df['latitude'].unique()
```

```python
df['country_code'].unique()
df['country_code'] = DataCleaning.remove_strings_of_undesired_length(df, 'country_code', 2)
df['country_code'].unique()
```

```python
df['continent'].unique()
df['continent'] = DataCleaning.remove_non_alpha_values(df, 'continent')
df['continent'].unique()
df['continent'] = df['continent'].replace('eeEurope', 'Europe')
df['continent'] = df['continent'].replace('eeAmerica', 'America')
df['continent'].unique()
```

```python
df = df[(df.address != '') & (df.store_code != '')]
df.address.sort_values()
df.store_code.sort_values()
```
