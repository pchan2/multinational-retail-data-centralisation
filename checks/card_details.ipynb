```python
import sys
import os
 
# Getting the name of the directory where the this file is present.
current = os.path.dirname(os.path.realpath('card_details'))
 
# Getting the parent directory name where the current directory is present.
parent = os.path.dirname(current)
 
# Adding the parent directory to the sys.path.
sys.path.append(parent)

from data_extraction import DataExtractor
from data_cleaning import DataCleaning
from database_utils import DatabaseConnector
```

```python
pdf_file = 'https://data-handling-public.s3.eu-west-1.amazonaws.com/card_details.pdf'
df = DataExtractor().retrieve_pdf_data(pdf_file)
df
```

```python
df = DataCleaning.remove_null_values(df)
df
```

```python
df['card_provider'].unique()
df['card_provider'] = DataCleaning.remove_non_card_providers(df, 'card_provider')
df['card_provider'].unique()
```

```python
df['date_payment_confirmed'].sort_values()
df['date_payment_confirmed'] = DataCleaning.remove_non_datetime_values(df, 'date_payment_confirmed')
df['date_payment_confirmed'] = DataCleaning.format_datetime_values(df, 'date_payment_confirmed')
df = DataCleaning.remove_null_values(df)
df['date_payment_confirmed'].sort_values()
```

```python
df['expiry_date'].sort_values()
df['expiry_date'].unique()
df['expiry_date'] = DataCleaning.remove_alphanumerical_values(df, 'expiry_date')
df['expiry_date'].unique()
```

```python
# df['card_number'].sort_values()
df['card_number'].unique()
df['card_number'] = df['card_number'].astype(str)
df['card_number'].sort_values()
df['card_number'] = DataCleaning.remove_non_digit_values(df, 'card_number')
df['card_number'] = df['card_number'].replace('', 0)
df['card_number'] = df['card_number'].astype(int)
df['card_number'].unique()
df['card_number'].sort_values()
```

```python
# delete rows where card_number != 0
df = df[df.card_number != 0]
df
```
