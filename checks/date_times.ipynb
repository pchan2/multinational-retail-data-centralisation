```python
import sys
import os

import pandas as pd
 
# Getting the name of the directory where the this file is present.
current = os.path.dirname(os.path.realpath('card_details'))
 
# Getting the parent directory name where the current directory is present.
parent = os.path.dirname(current)
 
# Adding the parent directory to the sys.path.
sys.path.append(parent)

import yaml

from data_cleaning import DataCleaning
from database_utils import DatabaseConnector
```

```python
with open(f'../config.yaml', 'r') as file:
    config = yaml.safe_load(file)
df = pd.read_json(config['JSON_FILE'])
df
```

```python
df['timestamp'].unique()
df['timestamp'].sort_values()
df['timestamp'] = DataCleaning.remove_alphanumerical_values(df, 'timestamp')
df['timestamp'].sort_values()
```

```python
df['month'].unique()
df['month'] = DataCleaning.remove_non_digit_values(df, 'month')
df['month'].unique()
```

```python
df['year'].unique()
df['year'] = DataCleaning.remove_non_digit_values(df, 'year')
df['year'].unique()
```

```python
df['day'].unique()
df['day'] = DataCleaning.remove_non_digit_values(df, 'day')
df['day'].unique()
```

```python
df['time_period'].unique()
df['time_period'] = DataCleaning.remove_non_time_periods(df, 'time_period')
df['time_period'].unique()
```

```python
df['date_uuid'].unique()
df['date_uuid'].sort_values()
```

```python
df = df.drop_duplicates()
df = DataCleaning.remove_null_values(df)
df
```

```python
df = df[(df['timestamp'] != '') | (df['month'] != '') | (df['year'] != '') | (df['day'] != '')]
df
```

```python
DatabaseConnector().upload_to_db(df, 'dim_date_times')
```
